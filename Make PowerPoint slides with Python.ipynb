```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pptx import Presentation
from pptx.util import Inches
from io import BytesIO
```

```python
df = pd.read_excel("data.xlsx")
```

```python
prs = Presentation("Template.pptx")
```

```python
slide = prs.slides.add_slide(prs.slide_layouts[5])
slide.shapes.title.text = 'This is a title'
```

```python
sns.boxplot(data=df, x="Credit_Score", y="Num_of_Delayed_Payment", hue="Payment_of_Min_Amount", palette="Pastel1")
plt.ylim(0,30)
memfile = BytesIO()
plt.savefig(memfile)
```

```python
slide.shapes.add_picture(memfile, left=Inches(1.5), top=Inches(2.5), height=Inches(3))
```

```python
sns.violinplot(data=df, x="Credit_Score", y="Num_of_Delayed_Payment", hue="Payment_of_Min_Amount", palette="Pastel1")
plt.ylim(0,30)
memfile = BytesIO()
plt.savefig(memfile)
```

```python
slide.shapes.add_picture(memfile, left=Inches(7), top=Inches(2.5), height=Inches(3))
```

```python
text_box = slide.shapes.add_textbox(left=Inches(1), top=Inches(6), width=Inches(10), height=Inches(1))
text_box.text_frame.text = "This is the text content"
```

```python
prs.save('Slides.pptx')
```
